```python
import numpy as np
from scipy import linalg
from matplotlib import pylab as plt
%matplotlib inline
```

```python
# построим кубическую параболу
x = np.arange(0, 30, 0.1) # значения икс список от -10 до 10 с шагом 0.1
y = (np.sin(x / 5))*(np.exp(x / 10)) + 5*np.exp(-x / 2)
plt.plot(x, y)
plt.show()
```

```python
# формула которая считает значения этой сложной функции в определенной точке
def formulya(x): 
    x = float(x)
    return(np.sin(x / 5) * np.exp(x / 10)+(5 * np.exp(-x / 2)))

print 'Значения функции при разных икс:'
print 'при x = 0\t f(x)=', formulya(0)
print 'при x = 1\t f(x)=', formulya(1)
print 'при x = 4\t f(x)=', formulya(4)
print 'при x = 8\t f(x)=', formulya(8)
print 'при x = 10\t f(x)=', formulya(10)
print 'при x = 15\t f(x)=', formulya(15)
```

```python
# Считаем первый порядок w0 + w1x = f(x)
uno = [1,15]
a1 = np.array([
    [uno[0]**0, uno[0]**1], 
    [uno[1]**0, uno[1]**1]]
)
b1 = np.array([formulya(uno[0]), formulya(uno[1])])
z1 = np.linalg.solve(a1, b1)
print z1
print 'Проверка = \t',z1[0]+z1[1], z1[0]+15*z1[1]
print 'Должно быть:\t', formulya(1), formulya(15)

x1 = np.arange(0, 15, 0.1) 
y1 = z1[0]+z1[1]*x1
plt.plot(x1, y1)
plt.title(u'Первый порядок')
plt.show()

plt.plot(x, y)
plt.title(u'Исходный')
plt.show()
```

```python
# Считаем второй порядок w0 + w1x+w2x^2 = f(x)
dos = [1,8,15]
a2 = np.array(
    [[dos[0]**0, dos[0]**1,dos[0]**2],
     [dos[1]**0, dos[1]**1,dos[1]**2],
     [dos[2]**0, dos[2]**1,dos[2]**2]]
)
b2 = np.array([formulya(dos[0]),formulya(dos[1]), formulya(dos[2])])
z2 = np.linalg.solve(a2, b2)
print z2

x2 = np.arange(0, 15, 0.1) 
y2 = z2[0]+z2[1]*x2+z2[2]*x2*x2
plt.plot(x2, y2)
plt.title(u'Второй порядок')
plt.show()

plt.plot(x, y)
plt.title(u'Исходный\n')
plt.show()
```

```python
# Считаем третий порядок w0 + w1x3+w2x3^2+w3x3^3 = f(x3)
Tre = [1,4,10,15]
a3 = np.array(
    [[Tre[0]**0,Tre[0]**1,Tre[0]**2,Tre[0]**3],
     [Tre[1]**0,Tre[1]**1,Tre[1]**2,Tre[1]**3],
     [Tre[2]**0,Tre[2]**1,Tre[2]**2,Tre[2]**3],
     [Tre[3]**0,Tre[3]**1,Tre[3]**2,Tre[3]**3]]
)

b3 = np.array([formulya(Tre[0]),formulya(Tre[1]), formulya(Tre[2]),formulya(Tre[3])])
z3 = np.linalg.solve(a3, b3)
print z3

x3 = np.arange(0, 16, 0.1) # значения икс список от -10 до 10 с шагом 0.1
y3 = z3[0]+z3[1]*x3+z3[2]*x3**2+z3[3]*x3**3
plt.plot(x3, y3)
plt.title(u'Третий порядок')
plt.show()

plt.plot(x, y)
plt.title(u'Исходный')
plt.show()
```
