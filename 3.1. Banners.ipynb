```python
import numpy as np
import pandas as pd
import sklearn
print(np.__version__, pd.__version__, sklearn.__version__)
```

```python
check = pd.read_csv('data/checkins.dat', sep='|', skipinitialspace=True, header=[0], skiprows=[1], low_memory=False)
check.columns = ['id','user_id','venue_id','latitude','longitude','created_at']
print(check.shape)
check.head(10)
```

```python
check = check.dropna()
print(check.shape)
check.head()
```

```python
# Оставляем первые 10к элементов
check_mini = check.iloc[:100000]
print(check_mini.shape)
check_mini.head()
```

```python
from sklearn.cluster import MeanShift
# coord = check_mini_2[['latitude', 'longitude']]
coord = check[['latitude', 'longitude']]
to_ms = np.array(coord)
print(to_ms.shape,'\n', to_ms)
```

```python
%%time
MS = MeanShift(bandwidth=0.1)
MS.fit(to_ms)
```

```python
MS_labels = MS.labels_
MS_centers = MS.cluster_centers_
MS_labels_u = np.unique(MS_labels)
MS_N_clusters = len(MS_labels_u)
print('Labels:\n', MS_labels.shape, MS_labels)
print('Centers:\n', MS_centers.shape, MS_centers)
print('Number of estimated clusters: ', MS_N_clusters)
```

```python
result = coord.copy()
result['cluster'] = MS_labels
result.sort_values('cluster')
```

```python
gruppirovka =  result.groupby('cluster').count()
print(gruppirovka.shape)
gruppirovka.head()
```

```python
big_cl = gruppirovka[gruppirovka['latitude'] >15]
print(big_cl.shape)
big_cl.tail()
```

```python
moi_klastery = list(big_cl.index)
print(len(moi_klastery),moi_klastery)
```

Некоторые из получившихся кластеров содержат слишком мало точек — такие кластеры не интересны рекламодателям. Поэтому надо определить, какие из кластеров содержат, скажем, больше 15 элементов. Центры этих кластеров и являются оптимальными для размещения.

```python
LA = (33.751277, -118.188740)
MM = (25.867736, -80.324116)
LN = (51.503016, -0.075479)
AM = (52.378894, 4.885084)
BJ = (39.366487, 117.036146)
SD = (-33.868457, 151.205134)
```

```python
fin = pd.DataFrame(MS_centers, columns=['Clust_Lat','Clust_Long'])
fin['to_LA']= (fin['Clust_Lat']-LA[0])**2 + (fin['Clust_Long']-LA[1])**2
fin['to_MM']= (fin['Clust_Lat']-MM[0])**2 + (fin['Clust_Long']-MM[1])**2
fin['to_LN']= (fin['Clust_Lat']-LN[0])**2 + (fin['Clust_Long']-LN[1])**2
fin['to_AM']= (fin['Clust_Lat']-AM[0])**2 + (fin['Clust_Long']-AM[1])**2
fin['to_BJ']= (fin['Clust_Lat']-BJ[0])**2 + (fin['Clust_Long']-BJ[1])**2
fin['to_SD']= (fin['Clust_Lat']-SD[0])**2 + (fin['Clust_Long']-SD[1])**2
fin
```

```python
fin['min_dist'] = fin[['to_LA','to_MM','to_LN','to_LN','to_BJ','to_SD']].min(axis=1)
fin.sort_values('min_dist')
```

```python
answer = list(fin.sort_values('min_dist').iloc[0, [0,1]])
print(answer)
with open('answers/3.1. Banners_answer1.txt', 'w') as f_out:
    out = ' '.join(str(i) for i in answer)
    f_out.write(str(out))
```
